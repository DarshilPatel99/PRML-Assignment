## Q1:-  Normal distribution

```python
#mean and standard deviation for normal distribution
mu = 0
sigma = 1
```

```python
#100 samples of normal distribution
normal_distribution_100 = np.random.normal(mu,sigma,100)

#histogram of normal distribution
frequency, bins, _ = plt.hist(normal_distribution_100 ,bins = 30 ,density=True,label='sample friquency')

#pdf for normal distribution with mu = 0 and sigma = 1
pdf = (1/(sigma*(np.sqrt(2*np.pi))))*np.exp(-(bins-mu)**2 / (2*(sigma**2)))

#plot pdf    
plt.plot(bins,pdf,label = 'pdf')
      
plt.legend()
plt.show()
```

## Q2 :- Bi-variate Normal distribution

```python
# mean and covarience matrix for bi-variate Normal distribution
mean = np.array([0,0])
cov = np.array([[1, 0], [0, 1]])
```

```python
#100 samples

#x,y are two variable of bi-variate Normal distribution
x, y = np.random.multivariate_normal(mean, cov, 100).T

# 3d plot for bi-variate Normal distribution
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# H is hight of bar(freqrency)
H, xedges, yedges = np.histogram2d(x, y,bins=(np.linspace(-5,5,30), np.linspace(-5,5,30)) )

xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()

ax.bar3d(xpos, ypos, 0, 1, 1, H.ravel(), zsort='average')
plt.show()
```

```python
#10000 samples

#x,y are two variable of bi-variate Normal distribution
x, y = np.random.multivariate_normal(mean, cov, 100000).T

# 3d plot for bi-variate Normal distribution
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# H is hight of bar(freqrency)
H, xedges, yedges = np.histogram2d(x, y,bins=(np.linspace(-5,5,30), np.linspace(-5,5,30)) )

xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()

ax.bar3d(xpos, ypos, 0, 1, 1, H.ravel(), zsort='average')
plt.show()
```

```python
# 2d plot for bi-variate Normal distribution
#it is just like like top view of 3d plot 
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

## Q3:- Central Limit Theorem

```python
# range for uniform distrebution is a to b
a = 0
b = 1

# number of samples for each uniform distribution
n = 12

#mean and standard deviation for uniform distribution
mu = (a+b)/2
sigma = np.sqrt(((b-a)**2)/12)

#10000 normal distribution each have n samples
uniform_distribution = np.random.uniform(a,b,(10000,n))

#uniform distribution to normal distribution using CLT
normal_distribution = (np.array([sum(xi) for xi in uniform_distribution]) - n*mu) / (np.sqrt(n)*sigma)
```

```python
#mean and varience for normal distribution 
# ~ N(0,1)
print(np.mean(normal_distribution))
print(np.var(normal_distribution))
```

```python
#plot histogram
plt.hist(normal_distribution,bins = 30)
plt.show()
```

## Q4 :- Box-Muller method

```python
# range for uniform distrebution u1,u2 is a to b
a = 0
b = 1

#number of samples
n = 10000

#n samples of normal distribution
u1 = np.random.uniform(a,b,n)
u2 = np.random.uniform(a,b,n)

#uniform distribution to normal distribution using Box-Muller method
#n samples of normal distribution
n1 = np.sqrt( (-2)*np.log(u1) ) * np.cos(2*np.pi*u2)
n2 = np.sqrt( (-2)*np.log(u1) ) * np.sin(2*np.pi*u2)
```

```python
#mean and varience for normal distribution n1
# ~ N(0,1)
print("n1:")
print("mean :",np.mean(n1))
print("variance :",np.var(n1))

#mean and varience for normal distribution n2
# ~ N(0,1)
print("n2:")
print("mean :",np.mean(n2))
print("variance :",np.var(n2))
```
