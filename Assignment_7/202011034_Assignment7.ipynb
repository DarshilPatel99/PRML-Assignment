```python
import numpy as np
import pandas as pd
import random
from sklearn.datasets import load_iris
```

```python
# loade iris dataset
dataset = load_iris()
x_data = dataset['data']
y_data = dataset['target']
labels = dataset['target_names']
features = dataset['feature_names']
```

```python
# Calculate Accuracy 
def accuracy(cluster_labels, target_labels):
    correct_pred = np.where(cluster_labels == target_labels)[0]
    accuracy = (correct_pred.shape[0] / len(cluster_labels)) * 100
    return accuracy
```

## Fuzzy C-Means

```python
# Number of Clusters
k = 3
# Maximum number of iterations
MAX_ITER = 200
# Number of data points
n = len(x_data)
# Fuzzy parameter
m =  3 
```

```python
# initializing the membership matrix each row has one '1' at rendom position else are zero 
def init_membership_mat(): 
    membership_mat = np.zeros((n, k))
    for i in range(n):
        ind = random.randint(0,k-1)
        membership_mat[i,ind] = 1
    return membership_mat
```

```python
 # calculating the center of all n clusters and return list of centers where each center is ndarray
def calc_cluster_center(membership_mat):
    cluster_centers = []

    for i in range(k):
        # denominator
        cluster = membership_mat[:,i]
        cluster_m = np.power(cluster,m)
        deno = np.sum(cluster_m)

        # array of numerator
        temp_num = np.zeros_like(x_data)
        for j in range(n):
            temp_num[j] = cluster_m[j]*x_data[j]
        numo = np.sum(temp_num,axis=0)

        # center
        center = numo/deno
        cluster_centers.append(center)

    # return list of centers
    return cluster_centers
```

```python
# Updating the membership matrix
def update_membership_mat(membership_mat, cluster_centers): 
    p = 2/(m-1)

    # calculate membership value for each point to each cluster centroid
    for i in range(n):
        x = x_data[i]
        distances = [np.linalg.norm(x - cluster_centers[j]) for j in range(k)]
        for j in range(k):
            deno = sum([(distances[j]/distances[c])**p for c in range(k)])
            # update the value
            membership_mat[i,j] = 1/deno       
    return membership_mat
```

```python
# getting the clusters
def get_clusters(membership_mat): 
    cluster_labels = list()

    # high membership value denotes the cluster
    for i in range(n):
        idx = max((val, idx) for (idx, val) in enumerate(membership_mat[i]))[-1]
        cluster_labels.append(idx)
    return cluster_labels
```

```python
# fuzzy c means clustering algorithum returns clusters' centroids and clusters' labels
def fuzzy_c_means_clustering(): 
    # initialize memberaship metrix
    membership_mat = init_membership_mat()
    for _ in range(MAX_ITER):
        # calculate center of clusters 
        cluster_centers = calc_cluster_center(membership_mat)
        # update membership matrix
        membership_mat = update_membership_mat(membership_mat, cluster_centers)
        # get cluster labels
        cluster_labels = get_clusters(membership_mat)

    return cluster_labels, cluster_centers
```

```python
cluster_labels, cluster_centers = fuzzy_c_means_clustering()
print(cluster_labels)
print(cluster_centers)
```

```python
# show labels and accuracy
print("labels :",cluster_labels)
acc = accuracy(cluster_labels, y_data)
print("Accuracy:",acc)
```

## Agglomerative clustering algorithm

```python
def MyAgglomerativeClustering(data,n_clusters):
    # final labels
    labels = []
    # clusters will update for each iteration
    clusters = []
    # at beginning every data point is a cluster
    clusters = np.expand_dims(data, axis=1).tolist()

    # ends when number of clusters = n_clusters
    while n_clusters < len(clusters):
        distance = np.inf
        expand_cluster_ind = 0 
        remove_cluster_ind = 0
        
        # select two different clusters
        # cluster 1
        for cluster1_ind in range(len(clusters)):
            cluster1 = clusters[cluster1_ind]
            # cluster 2
            for cluster2_ind in range(cluster1_ind+1,len(clusters)):
                cluster2 = clusters[cluster2_ind]

                # find distance between two points(from different clusters) and get minimum distance
                for point1 in cluster1:
                    for point2 in cluster2:
                        # Check distance using single linkage and euclidean distance
                        dst_p1_p2 = np.linalg.norm(np.array(point1) - np.array(point2))
                        if dst_p1_p2 < distance:  
                            distance = dst_p1_p2
                            expand_cluster_ind = cluster1_ind
                            remove_cluster_ind = cluster2_ind

        # merge clusters
        clusters[expand_cluster_ind].extend(clusters[remove_cluster_ind])
        # remove the cluster which is merged in another cluster
        clusters.pop(remove_cluster_ind)

    # get labels from clusters(for iris dataset only)
    for sample in data.tolist():
        if sample in clusters[0]:
            labels.append(0)
        elif sample in clusters[1]:
            labels.append(1)
        else:
            labels.append(2)
    
    return labels
```

```python
AC_labels = MyAgglomerativeClustering(x_data,3)

# show labels and accuracy
print(AC_labels)
acc = accuracy(AC_labels, y_data)
print("Accuracy :",acc)
```
