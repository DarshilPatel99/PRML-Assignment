```
#useful librarys for this assigment
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#parameters for 1st distribution
mu1 = 5
sigma1 = np.sqrt(10)

#parameters for 2nd distribution
mu2 = 10
sigma2 = np.sqrt(15)

#number of samples
n = 1000
```

```
# list of samples of 1st distribution
l1 = []
# list of samples of 2nd distribution
l2 = []
# list of samples of total distribution
l = []

# select samples
for i in range(n):
  # P1 = 0.6 and P2 = 0.4 is given, so first of all a random number(m) from 0 to 1 is selected
  # if m is <= 0.6 then sample is taken from 1st distribution
  # else sample is taken from 2nd distribution
  m = np.random.rand()
  if m <= 0.6:
    x = np.random.normal(mu1,sigma1)
    l1.append(x)
  else:
    x = np.random.normal(mu2,sigma2)
    l2.append(x)
  l.append(x)
```

```
# select range of the distribution
min = np.min(l)
max = np.max(l)
rng = np.linspace(min,max,100)

#histogram of 1st normal distribution
plt.hist(l1 ,bins = 30 ,density=True,alpha=0.5)

#pdf of 1st normal distribution
pdf1 = (1/(sigma1*(np.sqrt(2*np.pi))))*np.exp(-(rng-mu1)**2 / (2*(sigma1**2)))

#plot 1st pdf 
plt.plot(rng,pdf1,label = 'P1(x)~N(5,10)')

#histogram of 2nd normal distribution
plt.hist(l2 ,bins = 30 ,density=True,alpha=0.5)

#pdf of 2nd normal distribution
pdf2 = (1/(sigma2*(np.sqrt(2*np.pi))))*np.exp(-(rng-mu2)**2 / (2*(sigma2**2)))

#plot 2nd pdf
plt.plot(rng,pdf2,label = 'P1(x)~N(10,15)')

plt.legend()
plt.show()
```

```
# standard deviation for all sample
sigma = np.std(l)

# bandwidth
h = sigma*((4/(3*n))**(1/5))

KDE = []

# calculate K(X)
for X in rng:
  ans = 0
  for xi in l:
    a = (1/(h*(np.sqrt(2*np.pi))))
    b = (-0.5)*(((X - xi)/h)**2)
    ans += a*np.exp(b)
  ans /= (n*h)
  KDE.append(ans)
```

```
# plot of final probability distribution function
plt.plot(rng,KDE)
plt.show()
```
